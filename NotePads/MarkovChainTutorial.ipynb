## Markov Chain Basics

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

A Markov chain is a way of representing a sequence of possible events in which the event that is likely to take place is only a product of what event previously happened.

<img src="Chains.JPG">

Figure one is a diagram that represents a simple markov chain. Each of the nodes ( a ,  b and c) represents a valid state that the system could be in , with the arcs showing the what the probability is of the next state will be for each of the previous states. Taking Node A as an example , there is a 30% chance that If the current state is A then the next state will be A again , and a 70% chance that the next state will be B.

We can represent the markov chain above with a weighted adjacency matrix shown below. Each of the rows of the matrix represent the node that the state was in, whilst the columns represent the node that the state will end up in.
For example in position [0,0] the number 0.3 appears, this shows if the current state is A there is a 30% chance that the next state will be A
This matrix is effectively a transformation matrix.

```python
MarkovChain = np.array([
    [0.3 , 0.7 , 0  ] , 
    [0   , 0.5 , 0.5],
    [0.2 , 0   , 0.8]
    ])
print(MarkovChain)
```

If we pre-multiply the matrix by a matrix containing [1,0,0] we get out the matrix [0.3 , 0.7  , 0] this matrix represents the fact that if we definitely start at A , there is a 30% chance and a 70% chance of ending up at b.  If we then multiply  it by the matrix for a second time we get the matrix [0.09 0.56 0.35] this represents the probability that the state is each of the nodes after two steps.

```python
StartingPosition = np.array([1,0,0])
FirstPosition = np.matmul(StartingPosition,MarkovChain) # Matrix multiplying the starting position vector by the markov Chain Vector.
print("First Position" ,FirstPosition)
SecondPosition =np.matmul(FirstPosition, MarkovChain) # Matrix multiplying the first position vector by the markov Chain Vector.
print("Second Position" ,SecondPosition)
```

If you take the Markov matrix and multiply itself an infinite times ( or in our case in order to get a reasonably accurate approximation 100 times) , you get the steady state matrix. This matrix represents the what state the system is likely to be on when starting from any of the 3 starting point.
As shown by the fact that down each column of the matrix the value is the same. no matter on what state you begin on after taking an infite number of steps, the probability of ending up on one of the states is the same.

```python
StationaryPoint = np.linalg.matrix_power(MarkovChain , 100) # Multiplting the Markov chain vector by itsefl 100 times
print(StationaryPoint)
```

There is another way of calculating the stationary state, and that is using the perron eigenvector. As proven in various other places , for a positive matrix , there will always be one large positive eigenvalue,  and the vector related to this eigen value ,  for row a stochastic matrix at least will be the proportional to a row of the stationary state matrix.
As shown by the code below.


```python
#By defult numpy calculates the right hand eigen values , so the matrix has to be transposed before the eigen values and vectors are calculated
Values , Vectors = np.linalg.eig(np.transpose(MarkovChain))


print(Values)
PerronVector = Vectors[:,2]# From the printout it can be seeen that the perron value is the third value and so the third slice of the matrix is taken as the perron vector.
print(PerronVector)
NormalisingConstant = np.sum(PerronVector) #to make the perron vector a unit vector , a normalising constant is calculated by summing together the parts.
print(NormalisingConstant)
NormalisedPerron = PerronVector/ NormalisingConstant # the normalised perron vector is calculated by dividing the perron vector by the Normalising constant.
print(NormalisedPerron)
```
